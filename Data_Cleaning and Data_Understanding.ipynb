```python
# In the Data cleaning part pandas library is most preferable
import pandas as pd
import numpy as np
```

```python
#Impoting the file which we are done in scraping part using read_excel method
df = pd.read_excel('df.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
# when checking the data set it contain unnamed column and we drop that column with using drop() method
df.drop(columns = 'Unnamed: 0',inplace = True)
df.head()
```

#### In the Location column we have (), extra so we are removed that special character using strip method

```python

df['Location']=df['Location'].str[1:]
```

```python
df.head()
```

#### In the User_Comment column to delete (\n) again we use strip() method

```python
df['User_Comment']=df['User_Comment'].str[3:]
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

### Observations:

- The shape of the dataset 11980 rows and 8 columns.
- Presenting the data set having the column names as 
        - Bank_Name: Name of the banks in india
        - Bank_Type: Type of the bank account
        - Location: City of the Banks
        - Bank_Caption: Motto line of every bank
        - User_Name: name of the person who are given a rating for the bank
        - Bank_Rating: Rating of the bank by the user_name
        - Rating_Comment: Commenting in a single line by the username
        - User_Comment: it contains the experience of the User_Name
- At last we observe that all the columns are having data type as object except Bank_Rating column, it is a float data type

```python
# the data set contains some duplicate oservations so we removed those duplicates with using drop_duplicates()
df.drop_duplicates(inplace = True)
df.shape
```
