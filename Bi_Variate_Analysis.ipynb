```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Clean_data_set.csv')
```

```python
df.drop(columns = 'Unnamed: 0',inplace = True)
```

```python
df.tail()
```

```python
df.keys()
```

## Bank_Name(HDFC Bank) with Bank_Type, Location, Bank_Rating, Rating_Comment

```python
d1 = df[(df['Bank_Name']=='HDFC Bank')]
```

```python
d1['Bank_Type'].value_counts().plot.bar()
```

**- `Personal Loan Account` mode is maximum in the `HDFC Bank`**

**- It counts more than 1200 in the dataset**

```python
d2=d1['Location'].value_counts()
```

```python
f,ax = plt.subplots(1,2,figsize = (10,5)) # defining plot size and subplots
f.suptitle('Location vs count of values in Bank_Name(HDFC Bank)')
d2.head(20).plot('barh', ax = ax[0])
d2.tail(20).plot('bar',ax = ax[1])
```

**- Most of the `HDFC Banks` are located in Banglore, and the count is in between 270-300**

**- This bank is located in hyderabad i.e, second most frequent location count is 250-270**

```python
d3 = d1['Bank_Rating'].value_counts()
```

```python
f,ax = plt.subplots(1,2,figsize = (10,5)) # defining plot size and subplots
f.suptitle('Bank_Rating vs count of values in Bank_Name(HDFC Bank)')
d3.head(20).plot('barh', ax = ax[0])
d3.tail(20).plot('bar',ax = ax[1])
```

**- `HDFC Bank` rated mostely as 5.0, the count of 5.0 rating is near to 1000**

**- it is rated 4.0 as second most frequent count of approximately 700** 

**Given 1.5 rating as less number of times**

```python
d4 = d1['Rating_Comment'].value_counts()
```

```python
f,ax = plt.subplots(1,2,figsize = (10,5)) # defining plot size and subplots
f.suptitle('Rating_Comment vs count of values in Bank_Name(HDFC Bank)')
d4.head(20).plot('barh', ax = ax[0])
d4.tail(20).plot('bar',ax = ax[1])
```

**HDFC Bank quoted with a comment as `Blown` most number of times**

**Given `Pretty` as less number of times**
